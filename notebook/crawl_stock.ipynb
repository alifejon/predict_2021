```python
import pandas as pd
import os
import FinanceDataReader as fdr
```

```python
path = '../data'
list_name = 'stock_List.csv'
stock_list = pd.read_csv(os.path.join(path,list_name))
stock_list['종목코드'] = stock_list['종목코드'].apply(lambda x : str(x).zfill(6))
stock_list
```

```python
start_date = '20200101'
end_date = '20201231'
sample_code = '005930'
stock = fdr.DataReader(sample_code, start = start_date, end = end_date)
stock['code'] = sample_code
stock
```

```python
def read_data(code):
    stock = fdr.DataReader(code, start = start_date, end = end_date)
    stock['code'] = code
    return stock
stock_data = pd.concat(stock_list['종목코드'].apply(lambda x: read_data(x)).values.tolist())
stock_data
```
